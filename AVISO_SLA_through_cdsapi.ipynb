# 曾经，卫星高度计海平面数据集（如海平面高度、海浪等）主要通过​​法国国家太空研究中心（CNES）的**AVISO**高度计数据平台​​分发，自2015年和2016年起，已由哥白尼海洋环境监测服务（CMEMS）​​和​​哥白尼气候变化服务（C3S）​​接管。
# 需要注册ECMWF账号(和Copernicus账号不互通)
- 链接：`https://cds.climate.copernicus.eu/datasets/satellite-sea-level-global?tab=download`
- 通过右上角登陆/注册



```python
import cdsapi

year_list = [
        "2001", "2002", "2003",
        "2004", "2005", "2006",
        "2007", "2008", "2009",
        "2010", "2011", "2012",
        "2013", "2014", "2015",
        "2016", "2017", "2018",
        "2019", "2020"
    ]

dataset = "satellite-sea-level-global"
request = {
    "variable": ["monthly_mean"],
    "year": year_list,
    "month": [
        "01", "02", "03",
        "04", "05", "06",
        "07", "08", "09",
        "10", "11", "12"
    ],
    "version": "vdt2024"
}
for year in year_list:
     print('data in {} year is downloading...'.format(year))
     #每年一个文件，防止爆内存
     request["year"] = year
    #  target = r'D:\AVISO_SLA\SLA_{}'.format(year)
     target = r'D:\AVISO_SLA\SLA_2024_version\SLA_{}'.format(year)
    
     client = cdsapi.Client()
     client.retrieve(dataset, request, target)
     print('data in {} year has been downloaded'.format(year))
```
