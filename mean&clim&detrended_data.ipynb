```python
import numpy as np
import pandas as pd 
import xarray as xr
%matplotlib qt 
%pwd
```

```python
def load_data(which, where):
    filename = r'D:\all_processed_data\{}_ncdata\{}.nc'.format(where, which)
    return xr.open_dataarray(filename).rename(which)

def mean_data(data):
    return data.mean(dim='time')

def clim_data(data):
    time = pd.date_range('2001-01', '2020-12', freq='MS')
    year, month = time.year, time.month
    #将time*lon*lat转为year*month*lon*lat
    da = data.assign_coords(year=('time', year), month=('time', month))
    da = da.set_index(time=('year', 'month'))
    da = da.unstack('time')
    da = da.mean(dim='year')
    return da

def detrended_data(data):
    time = pd.date_range('2001-01', '2020-12', freq='MS')
    year, month = time.year, time.month
    data0 = clim_data(data)
    data = data.assign_coords(year=('time', year), month=('time', month))
    data = data.set_index(time=('year', 'month'))
    data = data.unstack('time')
    data = data-data0
    #将year*month*lon*lat转回time*lon*lat
    data = data.stack(time=('year', 'month'))
    data = data.drop_vars(['year', 'month', 'time'])  ##
    data = data.assign_coords(time=time)
    return data

which_list = [*list('SUVT')]
where_list = ['surface', '100m']
where_variable_list = [[], []]
for i, where in enumerate(where_list):
    for j in which_list:
        data = load_data(j, where)
        where_variable_list[i].append(mean_data(data)) 
        where_variable_list[i].append(clim_data(data))
        where_variable_list[i].append(detrended_data(data))


surface_data_var_dict = dict(mean_surface_S=where_variable_list[0][0], clim_surface_S=where_variable_list[0][1],
                     detrended_surface_S=where_variable_list[0][2], mean_surface_U=where_variable_list[0][3], 
                     clim_surface_U=where_variable_list[0][4], detrended_surface_U=where_variable_list[0][5],
                      mean_surface_V=where_variable_list[0][6], clim_surface_V=where_variable_list[0][7],
                     detrended_surface_V=where_variable_list[0][8],  mean_surface_T=where_variable_list[0][9], 
                     clim_surface_T=where_variable_list[0][10], detrended_surface_T=where_variable_list[0][11],
                     )

data_var_dict = dict(mean_100m_S=where_variable_list[0][0], clim_100m_S=where_variable_list[0][1],
                     detrended_100m_S=where_variable_list[0][2], mean_100m_U=where_variable_list[0][3], 
                     clim_100m_U=where_variable_list[0][4], detrended_100m_U=where_variable_list[0][5],
                      mean_100m_V=where_variable_list[0][6], clim_100m_V=where_variable_list[0][7],
                     detrended_100m_V=where_variable_list[0][8],  mean_100m_T=where_variable_list[0][9], 
                     clim_surface_T=where_variable_list[0][10], detrended_surface_T=where_variable_list[0][11],
                     )

surface_data = xr.Dataset(surface_data_var_dict, attrs=dict(description='表层S,U,V,T的平均态，气候态和去除气候态异常数据'))
surface_data.to_netcdf(r'D:\all_processed_data\surface_mean&clim&detrended_data\surface_data.nc')

hundred_data = xr.Dataset(surface_data_var_dict, attrs=dict(description='100m深S,U,V,T的平均态，气候态和去除气候态异常数据'))
hundred_data.to_netcdf(r'D:\all_processed_data\100m_mean&clim&detrended_data\hundred_data.nc')



```
